# Extraer solo categorias de interes

```python
categorias_deseadas = ["España", "Economía", "Cultura", "Deportes", "Ciencia", "Opinión"]  # Categorías que quieres filtrar
```

```python
# Realizar la solicitud HTTP a la página principal de El País
r = requests.get("https://www.elmundo.es/menu.html")

# Crear un objeto BeautifulSoup para analizar el HTML
soup = BeautifulSoup(r.text, "html.parser")

# Obtener todos los elementos <div> con la clase "col-mapweb"
columnas = soup.find_all("ul", class_="seccion-super-menu")

categorias = []
enlaces = []

# Encontrar todos los elementos li dentro de la lista ul
for col in columnas:
    
    enlaces_lista = col.find_all("a")
    for enlace in enlaces_lista:
        # Obtener el texto del enlace (subcategoría)
        categoria = enlace.text.strip()
        if categoria in categorias_deseadas:
        # Obtener el enlace (URL)
            url = enlace["href"]
            categorias.append(categoria)
            enlaces.append(url)
```

```python
# Buscar todos los párrafos dentro del div con la clase "a_c clearfix"
lista_categorias = []
lista_titulos = []
lista_noticias = []
lista_enlaces = []
lista_fechas = []
lista_autores = []  # Agregar lista para autores

for enlace, categoria in zip(enlaces, categorias):
    # Realizar una solicitud HTTP al enlace de la subcategoría
    print('ENLACE PRINCIPAL', enlace, categoria)
    r2 = requests.get(enlace)
    soup2 = BeautifulSoup(r2.text, "html.parser")

    # Obtener todos los artículos de la subcategoría
    articulos = soup2.find_all("article")

    # Obtener los enlaces de los artículos y obtener el texto de cada artículo
    for articulo in articulos:
        
        # Obtener el enlace del artículo
        link = articulo.find("a")["href"]
        
        titulo = articulo.find("h2").text.strip()
        # enlace_articulo = articulo.find("a")["href"]
        print(titulo)
        print(link)
        # Realizar una solicitud HTTP al enlace del artículo
        r3 = requests.get(link)
        soup3 = BeautifulSoup(r3.text, "html.parser")
        
        # Verificar si existe el div con la clase "a_c clearfix"
        div_articulo = soup3.find("div", class_="ue-c-article__body")
        # texto_articulo = soup3.find("div", class_="ue-c-article__body").get_text(strip=False)
        # texto_articulo = ""
        if div_articulo:
            texto_articulo = div_articulo.get_text(separator="\n")
        else:
            texto_articulo = ""
        
#         # Extraer la fecha
        fecha_element = soup3.find("div", class_="ue-c-article__publishdate")
        if fecha_element:
            fecha = fecha_element.find('time')['datetime']
        else:
            fecha = ""
            # autor_articulo = soup3.find('div', class_='ue-c-article__author-name-item').a.text
        author_element = soup3.find('div', class_='ue-c-article__author-name-item')
        if author_element:
            if author_element.a:  # Verificar si hay un enlace al autor
                autor_articulo = author_element.a.text
            else:
                autor_articulo = author_element.text.strip()
        else:
            autor_articulo = ""
        
        # Agregar los datos a las listas
        lista_categorias.append(categoria)
        lista_titulos.append(titulo)
        lista_noticias.append(texto_articulo)
        lista_enlaces.append(link)
        lista_fechas.append(fecha)  # Agregar la fecha
        lista_autores.append(autor_articulo)  # Agregar el autor

# Crear un DataFrame con los datos recopilados
df = pd.DataFrame({'categoria': lista_categorias, 
                   'enlace': lista_enlaces, 'titulo': lista_titulos,
                   'texto': lista_noticias, 'fecha': lista_fechas, 'autor': lista_autores})
```
