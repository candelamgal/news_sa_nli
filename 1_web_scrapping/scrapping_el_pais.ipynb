```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Extraer todas las categorias

```python
# Realizar la solicitud HTTP a la página principal de El País
r = requests.get("https://elpais.com/estaticos/mapa-web")

# Crear un objeto BeautifulSoup para analizar el HTML
soup = BeautifulSoup(r.text, "html.parser")

# Obtener todos los elementos <div> con la clase "col-mapweb"
columnas = soup.find_all("div", class_="col-mapweb")

# Listas para almacenar las categorías, subcategorías y enlaces
categorias = []
subcategorias = []
enlaces = []

# Iterar sobre cada columna
for col in columnas:
    # Obtener el título de la columna (categoría)
    categoria = col.find("h3").text.strip()
    
    # Encontrar todos los enlaces dentro de la lista
    url = col.find("a")['href']
    # Agregar la categoría, subcategoría y enlace a las listas correspondientes
    categorias.append(categoria)
    enlaces.append(url)
```

# Extraer solo categorias de interes

```python
categorias_deseadas = ["España", "Economía", "Cultura", "Deportes", "Ciencia", "Opinión"] 
```

```python
# Realizar la solicitud HTTP a la página principal de El País
r = requests.get("https://elpais.com/estaticos/mapa-web")

# Crear un objeto BeautifulSoup para analizar el HTML
soup = BeautifulSoup(r.text, "html.parser")

# Obtener todos los elementos <div> con la clase "col-mapweb"
columnas = soup.find_all("div", class_="col-mapweb")

# Listas para almacenar las categorías, subcategorías y enlaces
categorias = []
subcategorias = []
enlaces = []

# Iterar sobre cada columna
for col in columnas:
    # Obtener el título de la columna (categoría)
    categoria = col.find("h3").text.strip()
    
    # Encontrar todos los enlaces dentro de la lista
    url = col.find("a")['href']    
    if categoria in categorias_deseadas:# Iterar sobre cada enlace
        categorias.append(categoria)
        enlaces.append(url)
```

```python
# Buscar todos los párrafos dentro del div con la clase "a_c clearfix"
lista_categorias = []
lista_titulos = []
lista_noticias = []
lista_enlaces = []
lista_fechas = []
lista_autores = []  # Agregar lista para autores

for enlace, categoria in zip(enlaces, categorias):
    # Realizar una solicitud HTTP al enlace de la subcategoría
    print('ENLACE PRINCIPAL', enlace, categoria)
    r2 = requests.get(enlace)
    soup2 = BeautifulSoup(r2.text, "html.parser")

    # Obtener todos los artículos de la subcategoría
    articulos = soup2.find_all("article")

    # Obtener los enlaces de los artículos y obtener el texto de cada artículo
    for articulo in articulos:
        
        # Obtener el enlace del artículo
        link = articulo.find("h2").a["href"]
        if link.startswith("https://elpais.com/"):

            titulo = articulo.find("h2", class_="c_t").text.strip()
            # enlace_articulo = articulo.find("a")["href"]
            print(titulo)
            print(link)
            # Realizar una solicitud HTTP al enlace del artículo
            r3 = requests.get(link)
            soup3 = BeautifulSoup(r3.text, "html.parser")
            
            # Obtener el texto del artículo
            texto_articulo = ""
            
            # Verificar si existe el div con la clase "a_c clearfix"
            div_articulo = soup3.find("div", class_="a_c clearfix")
            if div_articulo:
                # Buscar todos los párrafos dentro del div
                for parrafo in div_articulo.find_all("p"):
                    texto_articulo += parrafo.get_text(separator="\n")
            else:
                texto_articulo = ""
            
            # Extraer la fecha
            fecha = soup3.find("div", class_="a_md_f").find(attrs={"data-date": True})["data-date"]
            
            # Extraer el autor
            div_autor = soup3.find("div", class_="a_md_a")
            if div_autor:
                autores = div_autor.find_all("a")
                nombres_autores = [autor.text.strip() for autor in autores]
                autor_articulo = ", ".join(nombres_autores)
            else:
                autor_articulo = "No especificado"
            
            # Agregar los datos a las listas
            lista_categorias.append(categoria)
            lista_titulos.append(titulo)
            lista_noticias.append(texto_articulo)
            lista_enlaces.append(link)
            lista_fechas.append(fecha)  # Agregar la fecha
            lista_autores.append(autor_articulo)  # Agregar el autor

# Crear un DataFrame con los datos recopilados
df = pd.DataFrame({'categoria': lista_categorias, 
                   'enlace': lista_enlaces, 'titulo': lista_titulos,
                   'texto': lista_noticias, 'fecha': lista_fechas, 'autor': lista_autores})
```
